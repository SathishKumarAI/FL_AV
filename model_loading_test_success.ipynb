```python

def load_model():
    # Instantiate YOLOv5 model
    model = YOLOV5(cfg="models/yolov5s.yaml")  # Assuming "models/yolov5s.yaml" is the path to the model configuration file
    # Load pre-trained weights
    pretrained_weights_path = os.path.join(os.getcwd()+os.sep+"yolov5s.pt") # Path to pre-trained weights file
    pretrained_dict = torch.load(pretrained_weights_path)
    # pretrained_dict["epoch"]
    pretrained_dict["model"]
    model.eval().model
    m_d = model.state_dict()
    pre= pretrained_dict["model"].state_dict()
    # Load the stm.eval().modelate_dict into the model
    csd = intersect_dicts(m_d,pre )
    # Optionally, move the model to GPU if available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.load_state_dict(csd, strict=False)
    model.to(device)
    return model
```
