```python
import os
import torch
import numpy as np
from ultralytics import YOLO
from collections import OrderedDict

# Model setup
MODEL_NAME = "yolov8s.pt"
SAVE_DIR = "client_models"
GLOBAL_MODEL_PATH = "models/global_yolov8.pt"
NUM_CLIENTS = 3  # Simulate 3 clients
NUM_ROUNDS = 5   # Simulate 5 federated learning rounds

os.makedirs(SAVE_DIR, exist_ok=True)
os.makedirs("models", exist_ok=True)


def get_weights(model: YOLO, save_path):
    """
    Extract YOLOv8 model weights and save to .pt file.
    """
    try:
        model.export(format="torchscript")  # Ensure model is compatible
        model.save(save_path)
        print(f" Model weights saved to {save_path}")
        return save_path
    except Exception as e:
        print(f" get_weights error: {e}")
        return None


def set_weights(model: YOLO, weights_path):
    """
    Load YOLOv8 weights from a .pt file.
    """
    try:
        if not os.path.exists(weights_path):
            print(f" Weight file {weights_path} not found!")
            return None

        model.model.load_state_dict(torch.load(weights_path)["model"], strict=False)
        print(f" Model updated from {weights_path} with strict=False")
        return model
    except Exception as e:
        print(f" set_weights error: {e}")
        return None

def load_weights_safely(model: YOLO, weights_path):
    """
    Load YOLOv8 weights safely by checking for mismatches.
    """
    try:
        state_dict = torch.load(weights_path)["model"]
        model_state_dict = model.model.state_dict()

        new_state_dict = OrderedDict()

        for name, param in model_state_dict.items():
            if name in state_dict and state_dict[name].shape == param.shape:
                new_state_dict[name] = state_dict[name]
            else:
                print(f" Skipping layer {name} due to shape mismatch: {param.shape} vs. {state_dict.get(name, 'MISSING').shape}")

        model.model.load_state_dict(new_state_dict, strict=False)
        print(" Model weights updated safely!")
        return model
    except Exception as e:
        print(f" Error loading weights safely: {e}")
        return None


def compare_weights(model1, model2):
    """
    Compare two YOLOv8 models' weights to ensure they match.
    """
    params1 = list(model1.model.parameters())
    params2 = list(model2.model.parameters())

    if len(params1) != len(params2):
        print(" Mismatch in number of parameters!")
        return False

    for i, (p1, p2) in enumerate(zip(params1, params2)):
        if not torch.allclose(p1, p2, atol=1e-6):
            print(f" Weight mismatch found at layer {i}!")
            return False

    print(" All weights match exactly!")
    return True


def federated_average(weight_files, save_path):
    """
    Perform simple federated averaging of multiple YOLOv8 models.
    :param weight_files: List of weight file paths (e.g., client weights)
    :param save_path: Path to save the averaged model.
    """
    models = [YOLO(w) for w in weight_files]  # Load all models
    params_list = [list(m.model.parameters()) for m in models]

    # Ensure all models have the same parameter structure
    num_params = len(params_list[0])
    for params in params_list:
        if len(params) != num_params:
            print(" Model parameter count mismatch!")
            return None

    # Federated averaging (element-wise mean)
    averaged_params = []
    for i in range(num_params):
        avg_param = torch.stack([params[i].detach().cpu() for params in params_list]).mean(dim=0)
        averaged_params.append(avg_param)

    # Load averaged weights into a new model
    global_model = YOLO(weight_files[0])  # Use first model as base
    state_dict = global_model.model.state_dict()
    new_state_dict = OrderedDict()

    for (name, param), avg_param in zip(state_dict.items(), averaged_params):
        new_state_dict[name] = avg_param.to(param.device)

    global_model.model.load_state_dict(new_state_dict, strict=False)
    global_model.save(save_path)
    print(f" Federated model saved to {save_path}")
    return global_model


# -----------------------------------------------
#  ITERATIVE FEDERATED LEARNING ROUNDS
# -----------------------------------------------
# Start with the global model
global_model = YOLO(MODEL_NAME)

for round_num in range(1, NUM_ROUNDS + 1):
    print(f"\n Federated Learning Round {round_num}/{NUM_ROUNDS}")

    client_weights = []

    # Simulate multiple clients training independently
    for client_id in range(1, NUM_CLIENTS + 1):
        print(f" Client {client_id} training...")

        client_model = YOLO(GLOBAL_MODEL_PATH if round_num > 1 else MODEL_NAME)  # Load global model

        # Simulate training by saving the model
        client_save_path = os.path.join(SAVE_DIR, f"client_{client_id}_round_{round_num}.pt")
        saved_file = get_weights(client_model, client_save_path)

        if saved_file:
            client_weights.append(saved_file)

            # Load and verify weight consistency
            loaded_model = set_weights(client_model, saved_file)
            if loaded_model:
                compare_weights(client_model, loaded_model)

    # Federated Averaging: Update global model
    if len(client_weights) >= 2:
        print("\n Updating global model via federated averaging...")
        # global_model = YOLO("yolov8s.pt")  # Load YOLOv8-Small

        global_model = federated_average(client_weights, GLOBAL_MODEL_PATH)

    print(f" Round {round_num} completed.\n")

print(" Federated Learning Simulation Finished!")
```
